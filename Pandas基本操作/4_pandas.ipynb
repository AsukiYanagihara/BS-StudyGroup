```python
"""
df.isnull() #欠損値をTrueで返す。.sum()で集計する
df.any() #欠損値のある行を抽出 axis=1?
df.fillna() #NaNを別のものに置き換える
df.dropna() #NaNがある行を消去　axis=0?
df.replace() #任意の指定したデータをなにかに置き換える
df.mask() #maskメソッド。よくわからん。
df.drop() #指定した列を削除。axis=1?
"""
```

```python
import pandas as pd
```

```python
df = pd.read_csv('lunch_box_3.csv')
```

```python
#列単位で欠損地Nan(not a number)が入っている個数をカウントする
#(正確には, isnull()でtrueが帰ってくる個数をカウント)
df.isnull().sum()
```

```python
#ひとつでもNaNが含まれる行だけを抽出(最初の5行のみ表示)
print(df[df.isnull().any(axis=1)].shape)
df[df.isnull().any(axis=1)].head()
```

```python
# 'payday'にあるNaNを0.0に置き換える
df.fillna(value={'payday': 0.0}, inplace=True)
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# kcal列にあるNanがある行を削除する
df.dropna(subset=['kcal'], axis=0, inplace=True)
print(df.shape) #207-166=41行のデータを消去
```

```python
df.isnull().sum()
```

```python
# 'precipitation'列の'--'を0に置き換える
df['precipitation'] = df['precipitation'].replace('--', 0).astype(float)
df.head()
```

```python
# maskメソッドを使う例。'sales'列が80より多ければ, その値を100に置換する
pd.DataFrame(df['sales'].mask(df['sales'] > 80, 100)).head()
```

```python
# 'remarks(備考)'はデータとして不要なので、データから列ごと削除
df.drop(['remarks'], axis=1, inplace=True)
df.head()
```

```python
df.to_csv('lunch_box_4.csv')
```
