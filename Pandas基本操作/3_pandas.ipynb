```python
# import
import pandas as pd
```

```python
# CSVの読み込み
df = pd.read_csv('lunch_box_2.csv')
```

```python
df.head()
```

```python
# datetimeをindexにする
df.set_index('datetime', inplace=True)
df.head()
# inplace=True >引数inplaceをTrueにすると、元のDataFrameが変更される。新しいDataFrameは返されず、返り値はNone。
```

```python
df.index
```

```python
# columns名を変更する。 (yをsalesにする)
df.rename(columns = {'y': 'sales'}, inplace=True)
df[100:110]
```

```python
# 'sales'列を降順で並び替える
df.sort_values(by="sales", ascending=True).head() #ascending=Trueで昇順
```

```python
df.sort_values(by="sales", ascending=False).head()
```

```python
# sort_valuesは複数の列の大しても実行できる。int, floatでなくてもOK
df.sort_values(['weather', 'sales'], ascending=True).head()
```

```python
# indexのデータタイプを確認
df.index #dtype='object'
```

```python
# indexであるdatetimeのdtype='object'からdtype='datetime64[ns]'に変更
df.index = pd.to_datetime(df.index, format='%Y-%m-%d')
df.index
```

```python
#indexがdatetimeのとくにできること
df.sort_index().head()
```

```python
# resampleメソッドで、日単位や月単位で簡単に集計する
df.resample('M').mean() #月単位で平均値を算出
```

```python
# 簡単にmonth列やday列を作れるようになる。
df['month'] = list(pd.Series(df.index).apply(lambda X: X.month))
df['days']=list(pd.Series(df.index).apply(lambda x : x.day))
```

```python
df.head()
```

```python
# cutメソッドでヒストグラフのビン指定的な処理ができる。
labels = ['上旬', '中旬', '下旬']
df['period'] = pd.cut(list(df['days']), bins=[0,10,20,31], labels=labels, right=True)# 0<day≦10, 10<day≦20, 20<day≦31
df.head()
#これはさすがにちょっとわからん。
```

```python
#あとで使うのでいったんCSVにアウトプット
df.to_csv('lunch_box_3.csv', encoding='utf-8')
```
