```python
import pandas as pd
```

```python
df = pd.read_csv('lunch_box_5.csv')
```

```python
df.set_index('datetime', inplace=True)
df.head()
```

```python
df.index = pd.to_datetime(df.index, format='%Y-%m-%d')
```

```python
#現在のcolumnsを確認
print(df.columns)
print(len(df.columns))
```

```python
#テキストデータが入っている列すべてをDummy化する
dummy_df = pd.get_dummies(df, columns=['week', 'name', 'event', 'weather', 'period'])
dummy_df.head()
```

```python
# jupyternotebookにおいて、pandasで表示が省略されるのを防ぐ
pd.set_option('display.max_columns', 161) #160番目まですべて強制表示
```

```python
print(dummy_df.shape)
dummy_df.head()
```

```python
#⑨最後にあらためて、各列にNanがはいってないか確認
print(dummy_df.isnull().sum())
dummy_df.head()
```

```python
# できたデータを保管
dummy_df.to_csv('processed_datamart.csv', encoding='utf-8')
```
