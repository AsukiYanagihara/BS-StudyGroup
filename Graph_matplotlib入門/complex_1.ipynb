```python
# モジュールのインポート
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#y軸の幅変更
x = np.arange(0, 10, 0.1)
y = np.exp(x)
plt.plot(x, y)
plt.title("exponetial function: $ y = e^x $") #Latexの数式
plt.ylim(0, 5000) #y軸スケール
```

```python
#複数のグラフを同一の領域に描写
xmin, xmax = -np.pi, np.pi
x = np.arange(xmin, xmax, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.hlines([-1, 1], xmin, xmax, linestyles="dashed") # y=-1, 1に破線を描写
plt.title(r"sin(x) and cos(x)") #rがわからん
plt.xlim(xmin, xmax) #xの範囲
plt.ylim(-1.3, 1.3) #yの範囲
```

```python
# グラフの分割
# xとyの値が上と同じ

#　sinxのプロット
plt.subplot(2, 1, 1) #plt.subplot(行数, 列数, 何番目のプロットか)
plt.plot(x, y_sin)
plt.title(r"$\sin x$")
plt.xlim(xmin, xmax)
plt.xlim(xmin, xmax) #xの範囲
plt.ylim(-1.3, 1.3) #yの範囲

#　cosxのプロット
plt.subplot(2, 1, 2) #plt.subplot(行数, 列数, 何番目のプロットか)
plt.plot(x, y_cos)
plt.title(r"$\cos x$")
plt.xlim(xmin, xmax)
plt.xlim(xmin, xmax) #xの範囲
plt.ylim(-1.3, 1.3) #yの範囲

plt.tight_layout() #タイトルのかぶりを防ぐ
```

```python
# グラフの分割
# xとyの値が上と同じ

#　sinxのプロット
plt.subplot(1, 2, 1) #plt.subplot(行数, 列数, 何番目のプロットか)
plt.plot(x, y_sin, "r") #b青 y黄色
plt.title(r"$\sin x$")
plt.xlim(xmin, xmax)
plt.xlim(xmin, xmax) #xの範囲
plt.ylim(-1.3, 1.3) #yの範囲

#　cosxのプロット
plt.subplot(1, 2, 2) #plt.subplot(行数, 列数, 何番目のプロットか)
plt.plot(x, y_cos, "g") #r赤くする g緑 b
plt.title(r"$\cos x$")
plt.xlim(xmin, xmax)
plt.xlim(xmin, xmax) #xの範囲
plt.ylim(-1.3, 1.3) #yの範囲
```
