```python
# モジュールのインポート
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#関数
x = np.linspace(0,10)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
#グラフ
plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label='cos')
plt.legend() #凡例をグラフにプロット
```

```python
#plt.legend()関数でarstistとラベルを対応づける
p1, = plt.plot(x, y_sin)
p2, = plt.plot(x, y_cos)
plt.legend([p1, p2], ["sin", "cos"])
```

```python
#関数部
x = np.random.randn(100)
y1 = x + np.random.randn(100)
y2 = x * 3 + np.random.randn(100) *2
```

```python
#グラフ
plt.plot(x, y1, "r.", label="y1") #. =dot
plt.plot(x, y2, "k.", label="y2") #. =dot
plt.legend()
```

```python
#グラフ
plt.plot(x, y1, "r.", label="y1") #. =dot
plt.plot(x, y2, "k.", label="y2") #. =dot
plt.legend(loc = "upper right") #右上に指定
```

```python
#グラフ
plt.plot(x, y1, "r.", label="y1") #. =dot
plt.plot(x, y2, "k.", label="y2") #. =dot
plt.legend(loc = "lower left") #左下に指定
```
