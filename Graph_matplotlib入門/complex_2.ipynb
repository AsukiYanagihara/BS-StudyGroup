```python
"""
やや複雑なグラフ
6個のグラフを描いてならべる
sin, cos, tan, sinh, cosh, tanh
"""
```

```python
# モジュールのインポート
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def plot_function_name(name, x=0, y=0):
    plt.text(x, y, name, alpha=0.3, size=25, ha="center", va="center")

x = np.arange(-3, 3.01, 0.01) # 3.01

plt.subplot(2,3,1)
plot_function_name("1: sin")
plt.plot(x, np.sin(x))

plt.subplot(2,3,2)
plot_function_name("2: cos")
plt.plot(x, np.cos(x))

plt.subplot(2,3,3)
plot_function_name("3: tan")
plt.plot(x, np.tan(x))
plt.ylim(-1,1) #y軸が無限になるので制限する

plt.subplot(2,3,4)
plot_function_name("4: sinh")
plt.plot(x, np.sinh(x))

plt.subplot(2,3,5)
plot_function_name("5: cosh", x=0, y=6) #graph nameの位置を調整
plt.plot(x, np.cosh(x))

plt.subplot(2,3,6)
plot_function_name("6: tanh")
plt.plot(x, np.tanh(x))
```
