```python
# モジュールのインポート
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# 関数の作成
n = 300
x = np.random.randn(n)
y1 = np.exp(x) + np.random.randn(n)
y2 = np.exp(x) * 3 + np.random.randn(n)
y3 = np.exp(x) + x
```

```python
#グラフを描く グラフ間でy軸が対応していない
fig = plt.figure()
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

ax1.plot(x, y1, ".")
ax2.plot(x, y2, ".")
ax3.plot(x, y3, ".")
```

```python
#グラフを描く グラフ間でy軸が対応していない
fig = plt.figure()
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132, sharey=ax1) #同じax1とy軸値
ax3 = fig.add_subplot(133, sharey=ax1) #同じax1とy軸値

ax1.plot(x, y1, ".")
ax2.plot(x, y2, ".")
ax3.plot(x, y3, ".")
```

```python
#別の種類のグラフでx軸を共有
fig = plt.figure()

#サブプロットを8:2で分割
ax1 = fig.add_axes((0, 0.2, 1, 0.8))
ax2 = fig.add_axes((0, 0, 1, 0.2), sharex=ax1)

#散布図のx軸のラベルとヒストグラムのy軸のラベルを非表示
ax1.tick_params(labelbottom="off")
ax2.tick_params(labelleft="off")

ax1.plot(x, y1, "x")
ax2.hist(x, bins=20) #bins・・・histの区間数
```
