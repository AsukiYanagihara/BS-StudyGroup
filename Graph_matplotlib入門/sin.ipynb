```python
"""
最も簡単な描画方法
sin関数を二次元平面に描画
散布図
"""
```

```python
# モジュールのインポート
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sin関数の描写
x = np.arange(-3, 3, 0.1) #x軸になる数字の列を作成.-3から3まで、0.1刻み
y = np.sin(x) #xに対応するy値
plt.plot(x, y) #関数の第一引数がx軸で第二引数がy軸
plt.plot(y, x)
```

```python
# 散布図
x = np.random.randn(30)
y = np.sin(x) + np.random.randn(30)
plt.plot(x, y, "o") # "o"は小さい円(circle marker) 
```

```python
# 散布図
x = np.random.randn(30)
y = np.sin(x) + np.random.randn(30)
plt.plot(x, y, "ro") # rはredの省略 
```
