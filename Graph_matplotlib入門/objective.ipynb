```python
"""
object指向でGraphを描く。
一体今何を操作してるのかが明確に
1. matplotlib.pyplot.figure関数でFigureオブジェクト
2. Figureオブジェクトのメソッド(.add_subplot, .add_axes)でAxesオブジェクトを作る。
3. xesオブジェクトのメソッド(.plot, .imshowなど)にデータを渡しプロットをする。
4. 必要ならば、Axesオブジェクトのメソッド(.set_ylimなど)でプロットを調節する。
"""
```

```python
# モジュールのインポート
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# 関数の作成
x = np.arange(-3, 3, 0.01)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
#グラフ描写
#1 Figureのインスタンス作成
fig = plt.figure()
print("type(fig): {}".format(type(fig))) #わからん

#2 Axesのインスタンス作成
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
print("type(fig): {}".format(type(ax1)))

#3 データを渡してプロット
ax1.plot(x, y_sin)
ax2.plot(x, y_cos)

#4 y軸範囲の調整とグラフタイトル・ラベル付け
ax1.set_ylim(-1.3, 1.3)
ax2.set_ylim(-1.3, 1.3)

ax1.set_title("sin x")
ax2.set_title("cos x")

ax1.set_xlabel("x")
ax2.set_xlabel("x")

ax1.set_ylabel("y")
ax2.set_ylabel("y")

fig.tight_layout() #かぶり
```
